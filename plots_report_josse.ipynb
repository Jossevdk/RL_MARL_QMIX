## IQL performance


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('IQL_CNN_win/score.csv')

# Column name
key = 'IQL_CNN_win__20241210_142411 - score'

# Calculate the rolling average and standard deviation
window_size = 100
df['Rolling_Avg'] = df[key].rolling(window=window_size, min_periods=1).mean()
df['Rolling_Std'] = df[key].rolling(window=window_size, min_periods=1).std()

# Plot the data with the rolling average and standard deviation as a spread
plt.figure(figsize=(6, 5))
plt.plot(df['Step'], df[key], label='Score')
plt.plot(df['Step'], df['Rolling_Avg'], label=f'Rolling Average ({window_size} steps)', color='red')
plt.fill_between(df['Step'], df['Rolling_Avg'] - df['Rolling_Std'], df['Rolling_Avg'] + df['Rolling_Std'], color='red', alpha=0.2, label='Rolling Std Dev')
plt.xlabel('Step')
plt.ylabel('Score')
plt.title('IQL versus random on win.lay: Score Over Steps')
plt.legend()
plt.grid(True)  # Add grid lines

plt.show()
```

```python
import pandas as pd
import numpy as np
# Read the first CSV file
df1 = pd.read_csv('IQL_CNN_blox_baseline/score1.csv')

# Read the second CSV file
df2 = pd.read_csv('IQL_CNN_blox_baseline/score2.csv')

array1 = df1[['Step', 'IQL_CNN_blox_baseline__20241210_203307 - score']].to_numpy()
array2 = df2[['Step', 'IQL_CNN_blox_baseline__20241210_210318 - score']].to_numpy()

array2[:,0] = array2[:,0] + array1[-1,0]
# Concatenate the DataFrames
array = np.concatenate((array1, array2), axis=0)
df = pd.DataFrame(array, columns=['Step', 'column'])

# Column name
key = 'column'

# Calculate the rolling average and standard deviation
window_size = 100
df['Rolling_Avg'] = df[key].rolling(window=window_size, min_periods=1).mean()
df['Rolling_Std'] = df[key].rolling(window=window_size, min_periods=1).std()

# Plot the data with the rolling average and standard deviation as a spread
plt.figure(figsize=(6, 5))
plt.plot(df['Step'], df[key], label='Score')
plt.plot(df['Step'], df['Rolling_Avg'], label=f'Rolling Average ({window_size} steps)', color='red')
plt.fill_between(df['Step'], df['Rolling_Avg'] - df['Rolling_Std'], df['Rolling_Avg'] + df['Rolling_Std'], color='red', alpha=0.2, label='Rolling Std Dev')
plt.xlabel('Step')
plt.ylabel('Score')
plt.title('IQL versus baseline on bloxCapture.lay: Score Over Steps')
plt.legend()
plt.grid(True)  # Add grid lines

plt.show()
```
