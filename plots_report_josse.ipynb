```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Read the CSV file
df = pd.read_csv('IQL_CNN_win/score.csv')
ep = np.arange(0, 401, 1)
# Column name
key = 'IQL_CNN_win__20241210_142411 - score'

# Calculate the rolling average and standard deviation
window_size = 100
df['Rolling_Avg'] = df[key].rolling(window=window_size, min_periods=1).mean()
df['Rolling_Std'] = df[key].rolling(window=window_size, min_periods=1).std()

# Plot the data with the rolling average and standard deviation as a spread
plt.figure(figsize=(6, 5))
plt.plot(ep, df[key], label='Score')
plt.plot(ep, df['Rolling_Avg'], label=f'Rolling Average ({window_size} steps)', color='red')
plt.fill_between(ep, df['Rolling_Avg'] - df['Rolling_Std'], df['Rolling_Avg'] + df['Rolling_Std'], color='red', alpha=0.2, label='Rolling Std Dev')
plt.xlabel('Episode')
plt.ylabel('Score')
plt.title('IQL versus random on win.lay: Score Over Steps')
plt.legend()
plt.grid(True)  # Add grid lines

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Read the CSV file
df = pd.read_csv('SimpleQMIX_CNN_random_blox_later_exploration/score.csv')
ep = np.arange(0, 301, 1)
# Column name
key = 'SimpleQMIX_CNN_random_blox_later_exploration__20241212_162015 - score'

# Calculate the rolling average and standard deviation
window_size = 100
df['Rolling_Avg'] = df[key].rolling(window=window_size, min_periods=1).mean()
df['Rolling_Std'] = df[key].rolling(window=window_size, min_periods=1).std()

# Plot the data with the rolling average and standard deviation as a spread
plt.figure(figsize=(6, 5))
plt.plot(ep, df[key], label='Score')
plt.plot(ep, df['Rolling_Avg'], label=f'Rolling Average ({window_size} steps)', color='red')
plt.fill_between(ep, df['Rolling_Avg'] - df['Rolling_Std'], df['Rolling_Avg'] + df['Rolling_Std'], color='red', alpha=0.2, label='Rolling Std Dev')
plt.xlabel('Episode')
plt.ylabel('Score')
plt.title('Qmix with later exploration versus random: Score Over Steps')
plt.legend()
plt.grid(True)  # Add grid lines

plt.show()
```
